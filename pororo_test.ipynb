```python
import pandas as pd
file_youtube=pd.read_csv("Youtube_Ko_Raw_04.csv")
file_news=pd.read_csv("news_keyword_04.csv")
file_twitter=pd.read_csv("twitter_cleaning_04.csv")
```

```python
from pororo import Pororo
ner=Pororo(task="ner",lang="ko")
review=Pororo(task="review_scoring",lang="ko")
```

```python
body_youtube=list(file_youtube['body'].head(10))
body_news=list(file_news['body'].head(10))
body_twitter=list(file_twitter['body'].head(10))
```

```python
from tqdm import tqdm_notebook
review_results=[]
for i in tqdm_notebook(body_news):
  result = review(i)
  review_results.append(result)
```
